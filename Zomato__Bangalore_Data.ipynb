```python
#importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df=pd.read_excel("/content/zomato_rest_data.xlsx")
df
```

2. Data Cleaning

```python
#Removing the unwanted columns --> Unnamed: 0,url,phone
df=df.drop(["Unnamed: 0","url","phone","address","reviews_list","menu_item","location","rest_type"],axis=1)
df
```

```python
#Cleaning name
df["name"]=df["name"].str.capitalize()
df["name"]
```

```python
#Cleaning rate column

#removing unwanted data
df["rate"] = df["rate"].str.strip("/5")

#Replacing the inconsistencies value with nan
df["rate"] = df["rate"].replace("NEW", np.nan)
df["rate"] = df["rate"].replace("-", np.nan)

#Changing the datatype
df["rate"]=df["rate"].astype(float)
df.info()

#Removing the nan values in rate column by filling the nan with mean
df["rate"]=df["rate"].fillna(df["rate"].mean())
df.isnull().sum()
```

```python
#Cleaning approx_cost(for two people)

#Removing the commas for the values present in approx_cost(for two people)
df["approx_cost(for two people)"]=df["approx_cost(for two people)"].str.replace(",","")

#Changing the datatype
df["approx_cost(for two people)"]=df["approx_cost(for two people)"].astype(float)
df.info()

#Handling the null values in approx_cost(for two people)
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].fillna(df["approx_cost(for two people)"].mean())
df.isnull().sum()
```

```python
#Renaming the listed_in(city) with location
df.rename(columns={"listed_in(city)": "location","listed_in(type)":"rest_type"}, inplace=True)
```

```python
#Handling null values in rest_type column
df["rest_type"]=df["rest_type"].fillna(df["rest_type"].mode)
df.isnull().sum()
```

```python
#Handling the null values in dishes liked
df["dish_liked"]=df["dish_liked"].fillna(df["dish_liked"].mode)
df.isnull().sum()
```

```python
#Handling the null values in cuisines
df["cuisines"]=df["cuisines"].fillna(df["cuisines"].mode)
df.isnull().sum()
```

#Task 2: Univariate Analysis

```python
# Top 5 locations with the highest number of restaurants
top_locations = df['location'].value_counts()

plt.bar(top_locations.index, top_locations.values, color=['red','gold'])
plt.title("Top 5 Restaurant Locations")
plt.xlabel("Location")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

**Interpretation:** The plot clearly shows that 'BTM' is the area with the most restaurants, followed by various 'Koramangala' blocks.


```python
# Distribution of restaurant ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['rate'], bins=20, kde=True, color='gold')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

**Interpretation:** The plot shows that most restaurant ratings fall between 3.5 and 4, indicating generally positive feedback. Lower ratings are less frequent.

```python
# Percentage of restaurants offering online orders
online_counts = df['online_order'].value_counts()

plt.pie(online_counts, labels=online_counts.index, autopct='%1.1f%%',colors=['red','gold'])
plt.title("Online Order Availability")
plt.show()
```

**Interpretation:** About **58.9%** of restaurants provide online ordering, showing that digital ordering is common. **41.1%** still do not offer online orders, indicating a significant portion relying on offline dining.

```python
# Distribution of average cost for two
avg_cost = df.groupby("location")["approx_cost(for two people)"].mean().sort_values(ascending=False)
sns.histplot(avg_cost, bins=10,color='red',kde=True)
plt.title("Locations by Average Cost for Two")
plt.xlabel("Location")
plt.ylabel("Average Cost")
plt.show()
```

**Interpretation**: Most locations have an average cost for two around **â‚¹450â€“â‚¹550**, which is where the distribution is most concentrated.A few locations have much higher costs **â‚¹650â€“â‚¹770**, creating the right-side tail in the distribution.

#TASK-3:Bivariate Analysis

```python
#Rating vs Cost
plt.scatter(df["rate"],df["approx_cost(for two people)"],alpha=0.3,color='red')

plt.title("Rating vs Cost for Two")
plt.xlabel("Cost for Two (â‚¹)")
plt.ylabel("Rating")
plt.show()
```

**Conclusion:** The majority of points are concentrated between â‚¹200 â€“ â‚¹1000, which means most restaurants fall in this price range. Regardless of price, most restaurants have ratings in this range. Even expensive restaurants â‚¹2000â€“â‚¹6000 do not consistently show better ratings than cheaper ones.There are only a small number of restaurants above â‚¹3000, which appear as scattered points on the right.

```python
#Rating vs Online Order
avg_rating = df.groupby("online_order")["rate"].mean().sort_values(ascending=False)
plt.bar(avg_rating.index, avg_rating.values,color=['red','gold'])
plt.title("Average Rating by Online Order Availability")
plt.xlabel("Online Order")
plt.ylabel("Average Rating")
plt.show()
```

**Conclusion:** The bar chart compares the average ratings of restaurants that offer online ordering with those that do not. Ratings is almost the same for both Online and Offline Order that means the customers like the restaurants equally and they might choose based on their convenience.

```python
# Votes vs Rating
plt.figure(figsize=(10, 6))
plt.scatter(df["votes"], df["rate"], alpha=0.3,color='red')

plt.title("Votes vs Rating")
plt.xlabel("Number of Votes")
plt.ylabel("Rating")
plt.show()
```

**Conclusion:** Most items have low number of votes, which is why many points appear on the left side of the graph. As the number of votes increases, the ratings are mostly high (around 4 to 5). This shows that popular items generally receive better ratings, while items with fewer votes have more varied ratings.
